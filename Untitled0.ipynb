# Resuelva gráficamente la función *f(x)=cos(x)-xe^x*

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
def fun(start, end, ylim=2, div=50):
  X = np.linspace(start, end, div)
  Y = np.cos(X) - X * np.exp(X)
  plt.plot(X,Y)
  plt.title(f"f(x) = cos(x) - xe^x   [{start}, {end}]")
  plt.grid()
  plt.axhline(0, color="black")
  plt.axvline(0, color="black")
  plt.xlim(start, end)
  plt.ylim(-ylim, ylim)
  plt.show()
```

```python
fun(-2,2)
```

```python
fun(0,1)
```

```python
fun(0.4, 0.6,0.4)
```

```python
fun(0.5,0.525, 0.025)
```

# Encuentre una raíz real de la ecuación *f(x) = 10x^2 − x − 4 = 0* correcta a dos dígitos significativos por el método de tabulación.

```python
def f(x):
  f = 10 * mt.pow(x, 2) - x - 4
  return f
```

```python
t1 = np.arange(0,1.1,0.1)
print("Tabulación 1")
print(f"x\tf(x)")
for i in t1:
  print(f"{round(i,1)}\t{round(f(round(i,1)),3)}")
print(f"\nTabulación 2")
t2 = np.arange(0.6,0.7,0.01)
print(f"x\tf(x)")
for i in t2:
  print(f"{round(i,2)}\t{round(f(round(i,2)),3)}")
print("Entonces, la raíz es 0.68")
```

# Encuentre una raíz real de la ecuación *f(x) = x^3 + x^2 − 10 = 0* correcta a dos dígitos significativos por el método de tabulación.

```python
def f(x):
  f = mt.pow(x, 3) + mt.pow(x, 2) - 10
  return f
```
